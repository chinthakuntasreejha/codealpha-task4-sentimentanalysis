```python
# 📌 TASK 4: SENTIMENT ANALYSIS
# Step 1: Install TextBlob
!pip install textblob
from textblob import TextBlob
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load Sample Amazon Review Dataset
url = "https://raw.githubusercontent.com/amankharwal/Website-data/master/Amazon%20Reviews.csv"
df = pd.read_csv(url)
df = df.rename(columns={'Reviews': 'review'})  # standardize column name
df = df[['review']].dropna().head(100)  # take only first 100 non-null reviews for demo
df.head()

# Step 3: Define Sentiment Function
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Step 4: Apply Sentiment Analysis
df['sentiment'] = df['review'].apply(get_sentiment)

# Step 5: Show Sample Results
print(df.head(10))

# Step 6: Countplot Visualization
sns.countplot(x='sentiment', data=df, palette='pastel')
plt.title("Sentiment Analysis on Amazon Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# Step 7: Save CSV File
df.to_csv("amazon_sentiment_analysis.csv", index=False)
```

```python
# 📌 Task 4: Sentiment Analysis (Self-contained Code)

# Step 1: Install TextBlob
!pip install textblob
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Create a manual list of Amazon-style reviews
reviews = [
    "I love this product, it's amazing and works perfectly!",
    "Terrible experience, I want a refund!",
    "It's okay, not great but not bad either.",
    "Absolutely fantastic! Exceeded my expectations.",
    "Not worth the price.",
    "Average quality, nothing special.",
    "Best purchase I've ever made.",
    "Awful, stopped working after a week.",
    "Pretty good, happy with it.",
    "Neutral opinion, nothing stood out."
]

# Step 3: Build DataFrame
df = pd.DataFrame({'review': reviews})

# Step 4: Define sentiment classification function
def get_sentiment(text):
    polarity = TextBlob(text).sentiment.polarity
    if polarity > 0.1:
        return "Positive"
    elif polarity < -0.1:
        return "Negative"
    else:
        return "Neutral"

# Step 5: Apply function
df['sentiment'] = df['review'].apply(get_sentiment)

# Step 6: Show results
print(df)

# Step 7: Plot results
sns.countplot(x='sentiment', data=df, palette='pastel')
plt.title("Sentiment Analysis on Sample Amazon Reviews")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

# Step 8: Save file
df.to_csv("task4_sentiment_analysis.csv", index=False)
```
